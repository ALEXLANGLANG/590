## Problem 5

```python
from scipy.io import loadmat
import numpy as np
from sklearn.metrics import mean_squared_error 
import matplotlib.pyplot as plt
import time
from mpl_toolkits import mplot3d
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
dataset = loadmat('dataset.mat')
D,X= dataset['D'],dataset['X']
```

### 5 (a)

```python
W_star = np.linalg.inv(X.T@X)@X.T@D   #get W directly
D_pred = X@W_star # get predicted D
Loss = mean_squared_error(D_pred,D)


print("Optimum W is: ")
print(W_star)
print("MES Loss of the whole dataset is: "+ str(Loss))
```

### 5 (b)

```python
#general function to train LMS model
def LMS_model(X,D,r, epochs = 20):
    W = np.array([0,0,0])
    W_list = []
    for j in range(epochs): #each epoch
        for i in range(100): #train each point one by one
            x,d = X[i,:],D[i]
            s = x@W.T
            W = W +  r*x*(d - s)
    
        D_pred = X@W # get predicted D
        Loss = 1/2*mean_squared_error(D_pred,D)
        W_list.append(Loss)
    return W_list, W

    
W_list,W =LMS_model(X,D,0.01) 
epochs = np.arange(1, 21, 1)
plt.title("MSE loss in log scale vs. Epochs")
plt.plot(epochs,np.log(W_list),ls = '-', marker='o')
plt.xlabel('Epochs')
plt.xticks(range(1,21))
plt.ylabel("log(MSE loss)")    
```

### 5 (c)

```python
def f(x, y,W):
    return W[0] + W[1]*x + W[2]*y 

fig = plt.figure( figsize = (8,7))

ax = plt.axes(projection='3d')
# Data for three-dimensional scattered points
zdata = D
xdata = X[:,1]
ydata = X[:,2]
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='hot');

x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)
X_, Y_ = np.meshgrid(x, y)
# model in (a)
Z = f(X_, Y_,W_star)
ax.contour3D(X_, Y_, Z, 50, cmap='gray')
# model in (b)
Z = f(X_, Y_,W)
ax.contour3D(X_, Y_, Z, 50, cmap='pink')

print(W)
print(W_star)
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('D')
```

The model in (a) is a gray plane and the model in (b) is a pink plane. From the piture above, we can see that these two planes almost overlap with each other and both of them fit the dataset well.

### 5 (d)

```python
r_list =[0.005, 0.01, 0.05, 0.5]
epochs = np.arange(1, 21, 1)


plt.figure(0)
plt.title("MSE loss in log scale vs. Epochs")
plt.xticks(range(1,21))
plt.ylabel("log(MSE loss)") 
plt.xlabel('Epochs')
time_list = []
for r in r_list:
    W_list, W = LMS_model(X,D,r)
    if r == 1:
        continue
    plt.plot(epochs,np.log(W_list),ls = '-', marker='o',label = 'r =' + str(r))
plt.legend()
plt.show()


plt.figure(1)
W_list, W = LMS_model(X,D,r= 1,epochs = 20)
plt.plot(epochs,np.log(W_list),ls = '--', marker='o',label = 'r =1 ')
plt.title("MSE loss in log scale vs. Epochs")
plt.xticks(range(1,21))
plt.ylabel("log(MSE loss)") 
plt.xlabel('Epochs')
plt.legend()

```

From the figure above, if the learning rate is very large like 1, it may be impossbile for us to get the optimum, which will have a bad impact on our training process. If r = 1, the model will diverge from the optimum and reach the infinity.  If r = 0.5, we cannot get the optimum either. 

If the learning rate is not too large to get a optimum, the smaller learning rate is, the less times of epochs we need to reach the optimum. In other words, the smaller learning rate we used, the more running time it takes.
