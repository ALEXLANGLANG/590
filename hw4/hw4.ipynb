### Lab 1, Part (b)

```python
def loss_1(inp_x, weights, label):
  return np.sum(np.square(np.matmul(inp_x, weights) - label))


X = np.array([np.array([1, -2, -1, -1, 1]), np.array([2, -1, 2, 0, -2]),  np.array([-1, 0, 2, 2, 1])])
Y = np.array([-7, -1, -1])
Y = Y.reshape(3,1)
weights = np.zeros((5,1))
u = 0.02
all_weights = []
loss = []
for j in range(200):
  grad = np.zeros((5,1))
  for i in range(3):
    X_i = X[i].reshape(1,5)
    grad = np.add(grad, (X_i * (Y[i] - np.matmul(X_i,weights))).T)
  weights = weights + (2*u *grad)
  # print(weights.shape)
  # print(X.shape)
  # print(Y.shape)
  all_weights += [weights]
  loss += [loss_1(X, weights, Y)]

print("Final weights for part b: ", weights)
print("Final Loss: \n", loss[len(loss)-1])
#print(np.asarray(all_weights))
plt.plot(range(200), np.log(np.asarray(loss)))
plt.xlabel("Step")
plt.ylabel("Log Loss")
plt.title("Log Loss as a function of step")
plt.show()

all_weights = np.array(all_weights)
plt.plot(range(200), all_weights[:,0], label = "w1")
plt.plot(range(200), all_weights[:,1], label = "w2")
plt.plot(range(200), all_weights[:,2], label = "w3")
plt.plot(range(200), all_weights[:,3], label = "w4")
plt.plot(range(200), all_weights[:,4], label = "w5")
plt.legend()
plt.xlabel("Step")
plt.ylabel("Weight Value")
plt.title("Weight Value as a function of step")
plt.show()



```

```python
head = 0
block_1 = 0
block_2 = 0
block_3 = 0
final = 0
for name, module in net.named_modules():
  #print(name)
  parser = name.split(".")
  if (len(parser)== 3 and parser[0] == "head_conv"):
    head += torch.numel(module.weight)
  if (len(parser) ==5):
    if (parser[1] == '0' or parser[1] =='1' or parser[1] =='2'):
      block_1 += torch.numel(module.weight)
    if (parser[1] == '3' or parser[1] =='4' or parser[1] =='5'):
      block_2 += torch.numel(module.weight)
    if (parser[1] == '6' or parser[1] =='7' or parser[1] =='8'):
      block_3 += torch.numel(module.weight)
  if (len(parser)== 2 and parser[0] == "final_fc"):
    final += torch.numel(module.weight)

print("Head: ", head)
print("Block_1: ", block_1)
print("Block_2: ", block_2)
print("Block_3: ", block_3)
print("Final: ", final)

```
