```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_base= pd.read_csv("acc_base.csv")
df_cutoout = pd.read_csv("acc_cutout.csv")
df_mixup = pd.read_csv("acc_mixup.csv")
```

```python
x = range(100)
model = [18,34,50]
for i in range(3):
    plt.plot(x,df_base.iloc[i,:],label = 'train-Resnet' + str(model[i]))
    plt.plot(x,df_base.iloc[i+1,:],label = 'test-Resnet' + str(model[i]))
    
    plt.plot(x,df_cutoout.iloc[i,:],label = 'Cutout: train-Resnet' + str(model[i]))
    plt.plot(x,df_cutoout.iloc[i+1,:],label = 'Cutout: test-Resnet' + str(model[i]))
    plt.legend()
    plt.show()
    
    plt.plot(x,df_base.iloc[i,:],label = 'train-Resnet' + str(model[i]))
    plt.plot(x,df_base.iloc[i+1,:],label = 'test-Resnet' + str(model[i]))
    plt.plot(x,df_mixup.iloc[i,:],label = 'mixup: train-Resnet' + str(model[i]))
    plt.plot(x,df_mixup.iloc[i+1,:],label = 'mixup: test-Resnet' + str(model[i]))
    plt.legend()
    plt.show()
```
